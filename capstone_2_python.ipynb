## Grouping National data by Decade.

```python
def make_decades(year):
    if 1900 <= year and 1909 >= year:
        return 1900
    elif 1910 <= year and 1919 >= year:
        return 1910
    elif 1920 <= year and 1929 >= year:
        return 1920
    elif 1930 <= year and 1939 >= year:
        return 1930
    elif 1940 <= year and 1949 >= year:
        return 1940
    elif 1950 <= year and 1959 >= year:
        return 1950
    elif 1960 <= year and 1969 >= year:
        return 1960
    elif 1970 <= year and 1979 >= year:
        return 1970
    elif 1980 <= year and 1989 >= year:
        return 1980
    elif 1990 <= year and 1999 >= year:
        return 1990
    elif 2000 <= year and 2009 >= year:
        return 2000
    elif 2010 <= year and 2019 >= year:
        return 2010
    else:
        return "pre-1900"

data_nat["Decade"] = data_nat['Year'].apply(make_decades)
data_nat1900 = data_nat[data_nat.Decade != "pre-1900"]
display(data_nat1900.head())

# for easier use in Tableau
data_nat1900_noF = data_nat1900.replace(to_replace = "F", value = "Female")
data_nat1900_noF = data_nat1900_noF.replace(to_replace = "M", value = "Male")

# saving the dataframe
# data_nat1900_noF.to_csv(r'C:\Users\jorda\Desktop\NationalNamesDecades_noF.csv')
```
